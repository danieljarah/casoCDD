## Random Forests

```python
# Definimos los vectores de características (X) y las etiquetas (y):
X_train = train_data.iloc[:, :-1] #exluye 'label'
y_train = train_data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=99)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(criterion = "gini",
                             max_depth=8,
                             min_samples_split= 10,
                             random_state=5)
```

```python
clf.fit(X_train, y_train)
```

```python
# Agregamos la data de test:
#X_test = test_data.drop(test_data.columns[1], axis=1)
#print(X_test)
```

```python
# Obtener predicciones
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```

## AdaBoost

```python
# Importamos el modelo
from sklearn.ensemble import AdaBoostClassifier

# Creamos el modelo con 50 estimadores
ada_clf = AdaBoostClassifier(n_estimators=50)
ada_clf = ada_clf.fit(X_train, y_train)

# Clasificamos el set de testeo y visualizamos
ada_pred = ada_clf.predict(X_test)

print('Performance metrics for AdaBoost:\n', classification_report(y_test, ada_pred))
```
